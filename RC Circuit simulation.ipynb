```python
import numpy as np
from scipy.integrate import odeint
from matplotlib import pyplot as plt

%matplotlib inline
```

Differential equation governing circuit
---

$$ \frac{dq}{dt} = -\frac{q}{RC} + \frac{V}{R} $$

```python
def dq_dt(q, t, params):
    R, C, V = params
    return (-q/(R*C) + V/R)
```

Parameters for circuit
---

```python
R = 5
C = 5
V = 5
Q0 = 0

params = (R, C, V)
```

```python
t = np.linspace(0, 100)

q = odeint(dq_dt, Q0, t, args=(params, ))
q = np.array(q).flatten()

i = dq_dt(q, t, params)
v = q/C
```

Plots
---

```python
plt.plot(t, q, label='Charge (Q)')
plt.plot(t, i, label='Current (I)')
plt.plot(t, v, label='Voltage (V)')
plt.legend()
```
